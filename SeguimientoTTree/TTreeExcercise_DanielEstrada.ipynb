```
import ROOT
```

```
from array import array
from numpy import random
```

```
MyTree = ROOT.TTree( 'tree1', 'tree para tres variables x(gaussianos),y(gaussianos),z(planos)')
```

```
mu_x, mu_y, sigma_x, sigma_y = 5,5,1,1

x = array( 'f', [0])
y = array( 'f', [0])
z = array( 'f', [0])
```

```
MyTree.Branch( 'X', x, 'X/F' )
MyTree.Branch( 'Y', y, 'Y/F' )
MyTree.Branch( 'Z', z, 'Z/F' )
```

```
for i in range(10000):
  x[0] = random.normal(mu_x,sigma_x)
  y[0] = round(random.normal(mu_y,sigma_y),2)
  z[0] = random.rand()*10
  MyTree.Fill()
```

```
c1 = ROOT.TCanvas( 'c1', 'My First Data example', 200, 10, 700, 900 )
c1.SetFillColor( 18 )
MyTree.Draw("X>>h1(20,0,10)")
c1.Update()
from ROOT import gROOT 
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("Y>>h2(20,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("Z>>h2(10,0,10)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("X*Y>>h3(50,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("X*Z>>h4(50,0,100)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```

```
MyTree.Draw("X*Y*Z>>h5(200,0,700)")
c1.Update()
gROOT.GetListOfCanvases().Draw()
```
